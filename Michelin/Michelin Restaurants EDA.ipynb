```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
one_star = pd.read_csv('C:/Users/User/Desktop/Oxford_DPhil/Kaggle/Michelin/MichelinRestaurants_data/one-star-michelin-restaurants.csv')
two_star = pd.read_csv('C:/Users/User/Desktop/Oxford_DPhil/Kaggle/Michelin/MichelinRestaurants_data/two-stars-michelin-restaurants.csv')
three_star = pd.read_csv('C:/Users/User/Desktop/Oxford_DPhil/Kaggle/Michelin/MichelinRestaurants_data/three-stars-michelin-restaurants.csv')
one_star['Michelin Stars'] = 1
two_star['Michelin Stars'] = 2
three_star['Michelin Stars'] = 3
r_data = pd.concat([one_star,two_star,three_star],axis=0).drop(columns=['url','zipCode'])
r_data = r_data[~r_data.city.isna()]
r_data
```

```python
plt.title('Missing Value Counts In Our Dataset',fontsize=20)
sns.heatmap(pd.DataFrame(r_data.isna().sum()).T,annot=True,cmap='Spectral',fmt='d',center=100)
sns.set(rc={'figure.figsize':(10,10)})
plt.show()
```

```python
plt.title('Top 10 Cities With The Most Michelin Restaurants',fontsize=20)
sns.barplot(x=r_data.city.value_counts()[:10],y=r_data.city.value_counts()[:10].index,palette='twilight')
plt.xlabel('Number Of Restaurants',fontsize=18)
plt.ylabel('Cities',fontsize=18)
plt.show()
```

```python
plt.title('Top 5 regions with the most michelin resto',fontsize=20)
sns.barplot(x=r_data.region.value_counts()[:5],y=r_data.region.value_counts()[:5].index,palette='Reds')
plt.xlabel('Number of resto',fontsize=20)
plt.ylabel('Regions',fontsize=20)
plt.show()
```

```python
region_number = r_data['region'].value_counts()

plt.title('top 5 regions with michelin',fontsize=20)
sns.heatmap(pd.DataFrame(region_number)[:5].T,cmap='Spectral',fmt='d'',center=100')
sns.set(rc={'figure.figsize':(10,3)})
plt.show()
```
